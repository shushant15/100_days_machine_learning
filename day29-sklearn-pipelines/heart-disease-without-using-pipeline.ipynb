```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart = fetch_openml(data_id=53, as_frame=True)
df = heart.frame
```

```python
heart_df = pd.DataFrame(heart.data , columns=heart.feature_names)
heart_df.isnull().sum()
heart_df.dtypes
```

```python
heart_df['target'] = heart.target
heart_df.head()
```

```python
# Test/Train Split

X_train, X_test, y_train, y_test = train_test_split(heart_df.drop(columns=['target']), heart_df['target'],test_size=0.2, random_state=42)
```

```python
X_train.head(5)
```

```python
y_train.sample(5)
```

```python
#As there is no missing values we wont able to apply imputers also all dtypes of columns are numerical as we are also not able to apply OHE or OE
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
acc = accuracy_score(y_test,y_pred)
print("Test Accuracy:", acc)
```

```python
import pickle

pickle.dump(lr,open('models/heart.pkl','wb'))
```
