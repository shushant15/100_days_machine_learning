```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression

```

```python
housing = fetch_california_housing()
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df.head()
```

```python
housing_df['target'] = housing.target
housing_df.head()
```

```python
housing_df.describe()
```

```python
housing_df.isnull().sum()
```

```python
housing_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(housing_df.drop(columns=['target']),housing_df['target'],test_size=0.2,random_state=42)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['AveRooms'])
plt.title('Avg Room')

plt.subplot(122)
stats.probplot(X_train['AveRooms'], dist="norm", plot=plt)
plt.title('Age QQ Plot')

plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['HouseAge'])
plt.title('House Age PDF')

plt.subplot(122)
stats.probplot(X_train['HouseAge'], dist="norm", plot=plt)
plt.title('House Age QQ Plot')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

clf =  LinearRegression()
clf2 = DecisionTreeRegressor()

clf.fit(X_train,y_train)
clf2.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = clf.predict(X_test)
y_pred1 = clf2.predict(X_test)

print("Linear Regression MSE:", mean_squared_error(y_test, y_pred))
print("Decision Tree MSE:", mean_squared_error(y_test, y_pred1))

print("Linear Regression R2:", r2_score(y_test, y_pred))
print("Decision Tree R2:", r2_score(y_test, y_pred1))
```

```python
from sklearn.preprocessing import FunctionTransformer

trf =  FunctionTransformer(func=np.log1p)

X_train_transformed = trf.fit_transform(X_train)
X_test_transformed = trf.transform(X_test)
```

```python
X_train_transformed.drop(columns=['Latitude', 'Longitude'], inplace=True)
X_test_transformed.drop(columns=['Latitude', 'Longitude'], inplace=True)
```

```python
X_train_transformed.head()
```

```python
clf =  LinearRegression()
clf2 = DecisionTreeRegressor()

clf.fit(X_train_transformed,y_train)
clf2.fit(X_train_transformed,y_train)
    
y_pred = clf.predict(X_test_transformed)
y_pred1 = clf2.predict(X_test_transformed)

print("Linear Regression MSE:", mean_squared_error(y_test, y_pred))
print("Decision Tree MSE:", mean_squared_error(y_test, y_pred1))

print("Linear Regression R2:", r2_score(y_test, y_pred))
print("Decision Tree R2:", r2_score(y_test, y_pred1))
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
stats.probplot(X_train['AveRooms'], dist="norm", plot=plt)
plt.title('AveRooms Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['AveRooms'], dist="norm", plot=plt)
plt.title('AveRooms After Log')

plt.show()
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
stats.probplot(X_train['HouseAge'], dist="norm", plot=plt)
plt.title('HouseAge Before Log')

plt.subplot(122)
stats.probplot(X_train_transformed['HouseAge'], dist="norm", plot=plt)
plt.title('HouseAge After Log')

plt.show()
```
